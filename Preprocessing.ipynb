### SIMPLE LINEAR REGRESSION

* import statsmodels.api for linear model

eq. of linear regression model: y = beta0 + beta1(x)

```python
import statsmodels.api as sn
```

by default statsmodels.api does not include beta0 value (often called constant). Therefore we will add it explicitly.

```python
X = sn.add_constant(df['room_num'])
```

this creates a constant(beta0) as well creates a predictor variable X.

```python
df.shape
```

```python
df.info()
```

```python
X
```

* to fit the model, we create object lm; OLS -> Ordinary least squares.

##### 1) OLS METHOD

```python
lm = sn.OLS(df['price'], X).fit()
```

```python
lm.summary()
```

* p value is preferred under 1% or 5% for better relationship & R-squared value should be inclined towards 1 i.e ~ 0.5 >= for better linear model.

To evaluate performance of our model we always look at the test dataset 

```python
# Standarize data before running Ridge & Lasso regression
```

```python
from sklearn import preprocessing 
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
# for ridge regression
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5) #here alpha is our lamda responsible for reducing coef of variables towards zero
```

```python
lm_r.fit(X_train_s, y_train)
```

```python
r2_score(y_test, lm_r.predict(X_test_s))
```

```python
# we will import validation curve because using that we can find out optimum r_2 value 
```

```python
from sklearn.model_selection import validation_curve
```

```python
# to find out the best value of lambda (here alpha) we run various models and choose the best optimum r2_score
```

```python
param_range = np.logspace(-2,8,100) #this logspace will create 100 values b/w 10^-2 & 10^8
```

```python
param_range
```

```python
# we will use these values for lambda and then choose best lambda value for model.
```

```python
train_scores, test_scores = validation_curve(Ridge(), X_train_s, y_train, "alpha", param_range, scoring = "r2" )
```

```python
# Above we have trained our model for all lambda values
```

* Observation: we haven't given test set for validation then how will we get test scores, it is because the test-train split is being done using Cross Validation or K-Fold validation not by validation set approach [which gives 80% (train_set) or 20% (test_set)]

```python
print(train_scores)
print(test_scores)
```

```python
# for each lambda we are getting 5 values of r2 as K-Fold validation is used.
```

```python
train_mean = np.mean(train_scores, axis = 1)
```

```python
test_mean = np.mean(test_scores, axis = 1)
```
